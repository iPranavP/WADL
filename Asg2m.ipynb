```python
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import classification_report
import tensorflow.keras.models as models
import tensorflow.keras.layers as layers
# import tensorflow.keras.optimizers as optimizers
import tensorflow.keras.datasets as datasets
# import tensorflow.keras.backend as backend
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv('/content/drive/MyDrive/Practice/MNIST(Ass2&3)/mnist_train.csv').values
test_data = pd.read_csv('/content/drive/MyDrive/Practice/MNIST(Ass2&3)/mnist_train.csv').values
```

```python
train_images = train_data[:,1:]
train_labels = train_data[:,0]

test_images = test_data[:,1:]
test_labels = test_data[:,0]
```

```python
print(train_images.shape)
print(train_labels.shape)

print(test_images.shape)
print(test_labels.shape)
```

```python
train_images = train_images/255;
test_images = test_images/255;
```

```python
model = models.Sequential()
model.add(layers.Dense(128,activation = 'sigmoid',input_shape = (28 * 28,)))
model.add(layers.Dense(64,activation = 'sigmoid'))
model.add(layers.Dense(10,activation = 'softmax'))
```

```python
model.compile(optimizer = 'sgd', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
history = model.fit(train_images,train_labels,epochs = 10, batch_size = 64, validation_split = 0.2)
```

```python
predictions = model.predict(test_images)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss')
plt.xlabel('Epoch')
plt.legend()
```

```python
plt.plot(history.history['accuracy'], label='Training Loss')
plt.plot(history.history['val_accuracy'], label='Validation Loss')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.legend()
```

```python
loss, accuracy = model.evaluate(test_images, test_labels)
```
